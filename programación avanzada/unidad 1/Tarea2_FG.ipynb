```python
#Visualizacin de Grficas 2D y 3D
#Autor: Felipe Gonzlez Boado

import numpy as np
import matplotlib.pyplot as plt
from funciones import Funcion

# definicin de valores funcin f
valores_f = np.linspace(-10, 10, 1000)

# definicin de valores funcin g
valores_x_g = np.linspace(-5, 5, 100)
valores_y_g = np.linspace(-5, 5, 100)

# evaluacin de valores
evaluacion_f = []
evaluacion_g = []

for valor in valores_f:
    f = Funcion(1, valor, 0, 0)
    evaluacion_f.append(f.evaluar())

for valor1 in valores_x_g:
    for valor2 in valores_y_g:
        g = Funcion(2, valor1, valor2, 0)
        evaluacion_g.append(g.evaluar())

# Grfica 2D de los valores de f(x) = x^2
plt.title("Grfica de $f(x) = x^2$", fontsize=21)
plt.xlabel("x")
plt.ylabel("y")
plt.plot(valores_f, evaluacion_f, label="$x^2$", lw=3)
plt.legend()
plt.savefig("grafica_f.png", dpi=300) #guardar la figura png
plt.show()

#Grfica 3D de los valores de g(x,y) = sen(x)cos(y)
X,Y = np.meshgrid(valores_x_g, valores_y_g)
Z = np.sin(X)*np.cos(Y)

import plotly.graph_objects as go
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y,
                                 colorscale = 'viridis',
                                 opacity = 0.9)])

fig.update_layout(title = 'Grfica de $g(x,y) = sen(x)cos(y)$',
                  scene = dict(xaxis_title = 'x',
                               yaxis_title = 'y',
                               zaxis_title = 'z'))

fig.write_html("grafica_g.html") # Guardar la grfica


fig.show()
```
